```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import keygen as kg
import numpy as np
def encryption(img, filename):
    #Generating chaotic keys
    height=img.shape[0]
    width=img.shape[1]
    key=kg.keygen(0.01,3.954,height*width)

    #Encryption by substituting with XOR
    z=0
    enimg= np.zeros(shape=[height,width], dtype=np.uint8)
    for i in range(height):
        for j in range(width):
            enimg[i,j]=img[i,j]^key[z]
            z+=1    
    plt.imsave(filename, enimg, cmap='gray')
    return enimg
```

```python
#Reading the image

img = cv2.imread('Images/Pappers/Pappers.tiff',0)
plt.imshow(img, cmap='gray')
plt.axis('off')
```

```python
def bitPlaneSlicing(r, bit_plane):
    dec = np.binary_repr(r,width = 8)
    return int(dec[8-bit_plane])
```

```python
bitPlaneSlicingVec = np.vectorize(bitPlaneSlicing)
```

```python
eighth_bitplane = bitPlaneSlicingVec(img, bit_plane = 8)
seventh_bitplane = bitPlaneSlicingVec(img, bit_plane = 7)
sixth_bitplane = bitPlaneSlicingVec(img, bit_plane = 6)
fifth_bitplane = bitPlaneSlicingVec(img, bit_plane = 5)
fourth_bitplane = bitPlaneSlicingVec(img, bit_plane = 4)
third_bitplane = bitPlaneSlicingVec(img, bit_plane = 3)
second_bitplane = bitPlaneSlicingVec(img, bit_plane = 2)
first_bitplane = bitPlaneSlicingVec(img, bit_plane = 1)
```

```python
cipher=(2 * (2 * (2 * (2 * (2 * (2 * (2 * enimg8 + enimg7) + enimg6) + enimg5) + fourth_bitplane) + third_bitplane) + second_bitplane) + first_bitplane);
plt.imshow(cipher, cmap='gray')
plt.imsave('Images/Pappers/cipher.tiff', cipher, cmap='gray')
```
